```python
import pandas as pd
from datetime import datetime
```

```python
def update_delete_transaction():
    with open("Expense&BudgetTracker.csv","r") as file:
        df=pd.read_csv("Expense&BudgetTracker.csv")
        print(df)
        t_id=input("Enter the transaction id which you want to delete/update")
        del_upd=input("Choose 1 for delete and 2 for update")
        text=file.read()
        file.seek(0)
        if(del_upd=='1'):
            for i in file.readlines():
                if t_id in i[0:4]:
                    new_text=text.replace(i,'')
            with open("Expense&BudgetTracker.csv","w") as file_new:
                file_new.write(new_text)
        if(del_upd=='2'):
            Amount=input("Enter the Expense/Income Amount")
            Date=input("Enter the Date in YYYY-MM-DD format")
            Comment=input("Enter the comment regarding expense/Income")
            Type=input("Enter Income/Expense")
            text=file.read()
            file.seek(0)
            new_text=str(t_id)+','+Date+','+Type+','+Amount+','+Comment+'\n'
            
            for i in file.readlines():
                if t_id in i[0:4]:
                    replaced_text=text.replace(i,new_text)
                    break
            print(i)
            print(new_text)
            print(replaced_text)
            with open("Expense&BudgetTracker.csv","w") as file_new:
                file_new.write(replaced_text) 
```

```python
def Add_income_expense():
    with open("Expense&BudgetTracker.csv","a+") as file:
        if(file.tell()==0):
            file.write("Date,Type, Amount, Comment") 
            Amount=input("Enter the Expense/Income Amount")
            Date=input("Enter the Date in YYYY-MM-DD format")
            Comment=input("Enter the comment regarding expense/Income")
            Type=input("Enter Income/Expense")
            input_text='1000'+Date+','+Type+','+Amount+','+Comment
            file.write(input_text)
            file.cl
        else:
            Amount=input("Enter the Expense/Income Amount")
            Date=input("Enter the Date in YYYY-MM-DD format")
            Comment=input("Enter the comment regarding expense/Income")
            Type=input("Enter Income/Expense")
            df=pd.read_csv("Expense&BudgetTracker.csv")
            transaction_id=int(df['Transaction_ID'].max())+1
            input_text="\n"+str(transaction_id)+','+Date+','+Type+','+Amount+','+Comment
            file.write(input_text)
```

```python
def view_transactions():
    with open("Expense&BudgetTracker.csv","r") as file:
        df=pd.read_csv("Expense&BudgetTracker.csv")
        print(df)
```

```python
def search_transaction():
    with open("Expense&BudgetTracker.csv","r") as file:
        date=input("Enter the transaction date")
        df=pd.read_csv('Expense&BudgetTracker.csv')
        df1=df[df['Date']==date]
        print(df1)
        if(df1['Date'].count()==0):
            print("No Transaction for the specified date is present")
search_transaction()
```

```python
def backup_file():
        with open('Expense&BudgetTracker.csv','r') as file:
            text=file.read()
            now = datetime.now()
            backup_file_name='Expense&BudgetTracker_backup_'+now.strftime("%Y%m%d_%H%M%S")+'.csv'
            with open(backup_file_name,'w') as f:
               f.write(text)


```

```python
def monthly_budget():
    with open('Expense&BudgetTracker.csv','r') as file:
        start_date='2025-08-01'#input('Enter the start date')
        end_date='2025-08-18'#input("Enter the end date")
        df=pd.read_csv("Expense&BudgetTracker.csv")
        filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
        income_df = filtered_df[filtered_df['Type'] == 'Income']
        expense_df = filtered_df[filtered_df['Type'] == 'Expense']
        print(f"\nExpenses for date range of {start_date} and {end_date} are")
        print(expense_df)
        print(f"\nIncomes for date range of {start_date} and {end_date} are")
        print(income_df)
        print("Summarized Income Date\n")
        print(filtered_df.groupby('Type').agg(Total_Amount=("Amount", "sum"),
                                                        Transaction_Count=("Amount", "count"),
                                                        Min_Transaction=("Amount", "min"),
                                                        Max_Transaction=("Amount", "max")))
```

```python
while(True):
    print("===== Expense & Budget Tracker =====")
    print("1. Add Expense\n\
    2. Add Income\n\
    3. View All Transactions\n\
    4. Search Transactions\n\
    5. Update/Delete Transaction\n\
    6. View Monthly Budget Summary\n\
    7. Set Monthly Budget\n\
    8. Generate Monthly Report (TXT)\n\
    9. Backup Data\n\
    10. Exit\n\
    ====================================")
    ch=input("Enter the choice")
    if(ch=="1" or ch=="2"):
        Add_income_expense()
    elif(ch=="3"):
        view_transactions()
    elif(ch=="4"):
        search_transaction()
    elif(ch=="5"):
        update_delete_transaction()
    elif(ch=="6"):
        monthly_budget()
    elif(ch=="7" or ch=="8"):
        print("Work In Progress")
        continue
    elif(ch=="9"):
        backup_file()
    elif(ch=="10"):
        break
```
